```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score, classification_report,r2_score, mean_squared_error
```

```python
#EDA
df=pd.read_csv("PB_All_2000_2021.csv", sep=";")
print("\n Sample of Table \n", df.head(10))
print("\n Info of Table \n")
df.info()
print("\n Shape of Table \n",df.shape)
print("\n Stastics of Table \n",df.describe())
print("\n Null characters of Table \n",df.isnull().sum())
```

```python
#Changing date format
df['date']=pd.to_datetime(df['date'], format='%d.%m.%Y')
df=df.sort_values(by=['id','date'])
df.head(10)
```

```python
# Removing date collumn and adding year, month and day
df['year']= df['date'].dt.year
df['month']= df['date'].dt.month
df['day'] = df['date'].dt.day
df=df.sort_values(by=['id','year','month','day'])
df.drop('date', axis=1, inplace=True)
df.head(10)
```

```python
# Reorder the DataFrame
cols = df.columns.tolist()
cols.insert(cols.index('id') + 1, cols.pop(cols.index('year')))
cols.insert(cols.index('id') + 2, cols.pop(cols.index('month')))
cols.insert(cols.index('id') + 3, cols.pop(cols.index('day')))
df = df[cols]
df.head()
```

```python
df.columns
```

```python
components_of_water=['NH4', 'BSK5', 'Suspended', 'O2', 'NO3', 'NO2', 'SO4','PO4', 'CL']
```

```python
df= df.dropna(subset=components_of_water)
```

```python
df.isnull().sum()
```

```python
#feature selection
X=df[['id','year']]
y= df[components_of_water]
```

```python
# enconding - one -hot 
X_encoded=pd.get_dummies(X,columns=['id'],drop_first=True)
X_encoded
```

```python
# Split the data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.25, random_state=42)
```

```python
# train model
model=MultiOutputRegressor(RandomForestRegressor(n_estimators=100,random_state=42))
model.fit(X_train,y_train)
```

```python
# model evaluation components_of_water
y_pred=model.predict(X_test)
```

```python
# Create lists to store results
mse_list = []
r2_list = []
```

```python
# Loop through each component
for i, col in enumerate(components_of_water):
    mse = mean_squared_error(y_test.iloc[:, i], y_pred[:, i])
    r2 = r2_score(y_test.iloc[:, i], y_pred[:, i])
    mse_list.append(mse)
    r2_list.append(r2)
```

```python
# Create a summary DataFrame
performance_df = pd.DataFrame({
    'Component': components_of_water,
    'MSE': mse_list,
    'R² Score': r2_list
})
performance_df
```

```python
# Display the table
print("Model Performance Summary:\n")
print(performance_df.to_string(index=False))
```

```python
station_id='22'
year_input=2022

input_data= pd.DataFrame({'year': [year_input],'id':[station_id]})
input_encoded= pd.get_dummies(input_data,columns=['id'])

# alingning with training feature columns
missing_cols=set(X_encoded.columns)- set(input_encoded.columns)
for col in missing_cols:
    input_encoded[col] =0
input_encoded = input_encoded[X_encoded.columns]

# predict water components
predicted_components= model.predict(input_encoded)[0]

print(f"\n Predicted components in water of water station '{station_id}' in {year_input} :" )
for p , val in zip (components_of_water,predicted_components):
    print(f" \t{p}: {val:.2f}")
```
